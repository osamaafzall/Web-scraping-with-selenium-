```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager
import time


driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))


driver.get("https://www.flipkart.com/")


try:
    close_button = WebDriverWait(driver, 5).until(
        EC.element_to_be_clickable((By.XPATH, "//button[contains(text(),'✕')]"))
    )
    close_button.click()
except:
    pass  


search_box = WebDriverWait(driver, 10).until(
    EC.presence_of_element_located((By.XPATH, "//input[@type='text']"))
)
search_box.send_keys("watch for women" + Keys.RETURN)

time.sleep(3)  
import pandas as pd


df = pd.DataFrame(columns=["Name", "Price"])


page = 1
while page <= 10:  
    print(f"Page {page}\n" + "="*20)

    
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(2)  

    
    products = driver.find_elements(By.XPATH, "//div[contains(@class,'syl9yP')]/parent::div/parent::div")

    for product in products:
        try:
            
            name = product.find_element(By.XPATH, ".//div[contains(@class,'syl9yP')]").text

            
            price = product.find_element(By.XPATH, ".//div[contains(@class,'Nx9bqj')]").text

            
            print(f"Name: {name}, Price: {price}")

            
            df.loc[len(df)] = [name, price]

        except:
            pass  

    
    try:
        
        next_button = WebDriverWait(driver, 5).until(
            EC.element_to_be_clickable((By.XPATH, "//a[span[text()='Next']]"))
        )
        next_button.click()
        time.sleep(5)  
    except:
        print("No more pages.")
        break  

    page += 1


df.to_csv('flipkart_women_watches.csv', index=False, encoding='utf-8')
print("Data has been written to flipkart_women_watches.csv")
```
